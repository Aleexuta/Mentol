```python
positive=pd.read_csv('../datasets/positive_reddit_combined.csv')
negative=pd.read_csv('../datasets/negative_reddit_combined.csv')
```

```python
from autocorrect import Speller
```

```python
spell = Speller(fast=True)
```

```python
def correct_sentence(row):
    return spell(str(row))
```

```python
positive['post']=positive['post'].apply(correct_sentence)
negative['post']=negative['post'].apply(correct_sentence)
```
